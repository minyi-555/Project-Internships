```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
from datetime import datetime
import pandas as pd
import time
import random
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
df = pd.read_csv('stock_code.csv')
df
```

```python
df['stock_code']=df['stock_code'].str.slice(0,-3)
```

```python
stock_code=list(df['stock_code'])
stock_code
```

```python
stock_name = []
stock_cat = []

# share price details
share_price = []
sp_per_chg = []
sp_chg = []
high = []
low = []
open_p = []
close_p = []
volume = []

profit = []
EPS = []
D_Y = []
ROE = []
ask = []
mar_cap = []
P_B = []
DPR = []
P_E = []
DPS = []
ex_div_date = []
bid = []
rating = []


driver = webdriver.Chrome()
driver.implicitly_wait(220)

# funtion to get link
def get_link(stock_code):
    
    url='https://www.bursamarketplace.com/index.php'
    driver.get(url)
    
    #click search button
    s= driver.find_element(By.XPATH,'//*[@id="newnav-mobileSearch"]')
    driver.execute_script("arguments[0].click();",s)
    
    # input code
    driver.find_element(By.XPATH, '//*[@id="newnav-search-input"]').send_keys(stock_code)
    
    # get the link
    lnks = driver.find_element(By.ID,'url')
    links=lnks.get_attribute('href')
    return links
    
#     content = driver.page_source
#     print(content)
#     soup = bs(content)
#     # search 
#     driver.find_element(By.XPATH, "//input[@name='newnav-search-input']").send_keys(stock_code)
    
stock_c=['7152','0025']
for s in stock_c:
    url=get_link(s)
    driver.get(url)
    driver.implicitly_wait(220) #maximum time to loadÂ theÂ link
    content = driver.page_source
    soup = bs(content)
    try:
        # share price details 
        stock_n= soup.find('div', attrs = {'name':'tixStockName', 'class':'topPnl_name'}) 
        stock_c = soup.find('span', attrs = {'name':'tixStockCat'}) 
        price = soup.find('div', attrs = {'name':'tixStockLast','class':'value'})
        #price_chp = soup.find('div', attrs = {'name':'tixStockPchg' ,'class':'pChg txt_down'})
        #price_ch = soup.find('div', attrs = {'name':'tixStockChg','class':'chg'})
        price_h = soup.find('div', attrs = {'name':'tixStockHigh','class':'value'})
        price_l = soup.find('div', attrs = {'name':'tixStockLow','class':'value'})
        price_open = soup.find('div', attrs = {'name':'tixStockOpen','class':'value'})
        price_close = soup.find('div', attrs = {'name':'tixStockClose','class':'value'})
        #price_vol = soup.find('div', attrs = {'name':'tixStockVolume','class':'value'})
    
        # Financial metric details
        stock_prof = soup.find('div', attrs = {'name':'tixStockProfit','class':'value'})
        stock_eps = soup.find('div', attrs = {'name':'tixStockEps','class':'value'})
        stock_d_y = soup.find('div', attrs = {'name':'tixStockDivYield','class':'value'})
        stock_roe = soup.find('div', attrs = {'name':'tixStockRoe','class':'value'})
        stock_ask = soup.find('div', attrs = {'name':'tixStockAsk','class':'value'})
        stock_mar_cap = soup.find('div', attrs = {'name':'tixStockMktCap','class':'value'})
        stock_p_b = soup.find('div', attrs = {'name':'tixStockPBook','class':'value'})
        stock_dpr = soup.find('div', attrs = {'name':'tixStockDivPayRate','class':'value'})
        stock_p_e = soup.find('div', attrs = {'name':'tixStockPe','class':'value'})
        stock_dps = soup.find('div', attrs = {'name':'tixStockDivPs','class':'value'})
        stock_ex_div_date = soup.find('div', attrs = {'name':'tixStockExDivDate','class':'value'})
        stock_bid = soup.find('div', attrs = {'name':'tixStockBid','class':'value'})
        stock_rating = soup.find('span', attrs = {'name':'tixStockAvgScore'})
    
    except:
        print('Information cannot scrap')
# Financial Health ratio
    
    
    try:
        # Append to respective list
        stock_name.append(stock_n.text)
        stock_cat.append(stock_c.text)
        share_price.append(price.text)
        #sp_per_chg.append(price_chp.text)
        #sp_chg.append(price_ch.text)

        high.append(price_h.text)
        low.append(price_l.text)
        open_p.append(price_open.text)
        close_p.append(price_close.text)
        #volume.append(price_vol.text)


        profit.append(stock_prof.text)
        EPS.append(stock_eps.text)
        D_Y.append(stock_d_y.text)
        ROE.append(stock_roe.text)
        ask.append(stock_ask.text)
        mar_cap.append(stock_mar_cap.text)
        P_B.append(stock_p_b.text)
        DPR.append(stock_dpr.text)
        P_E.append(stock_p_e.text)
        DPS.append(stock_dps.text)
        ex_div_date.append(stock_ex_div_date.text)
        bid.append(stock_bid.text)
        rating.append(stock_rating.text)
    
    except:
        print('Text cannot extract')
    # Program to generate a random number between 5 to 30
    # importing the random module
    sec = random.randint(5,30)

    # program scrapping waiting time
    time.sleep(sec) # Seconds


# 
df = pd.DataFrame({'Stock Name':stock_name, 'Stock Category':stock_cat, 'Share Price':share_price
                   , 'High':high,'Low':low, 'Open Price':open_p, 'Close Price':close_p,
                   'Profit':profit, 'EPS':EPS, 'DivYield':D_Y ,'ROE':ROE, 'Ask':ask, 
                   'Market cap':mar_cap,'P/B':P_B,'Div payoutR':DPR , 'P/E':P_E, 
                   'Div perS':DPS,'Ex. div date':ex_div_date,
                   'Bid':bid,'Rating':rating}) 
df.to_csv('Share.csv', index=False, encoding='utf-8')
```
